```python
model=YOLO('yolo11n.pt')
```

```python
project = rf.workspace("roboflow-universe-projects").project("license-plate-recognition-rxg4e")
version = project.version(11)
dataset = version.download("yolov8")
```

```python
model.train(data='/content/License-Plate-Recognition-11/data.yaml',epochs=1)
```

```python
import cv2
from google.colab.patches import cv2_imshow
from ultralytics import YOLO
# Load the video
input_path = "test_video.mp4"
cap = cv2.VideoCapture(input_path)

# Get video properties
fps = cap.get(cv2.CAP_PROP_FPS)
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Define the codec and create VideoWriter for AVI output
fourcc = cv2.VideoWriter_fourcc(*'XVID')  # or use 'mp4v' for .mp4
output_path = "annotated_output.avi"
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

# Process video frame-by-frame
while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break

    # Run YOLOv8 on the frame
    results = model(frame)

    # Plot detections on the frame
    annotated_frame = results[0].plot()

    # Write the annotated frame to the output video
    out.write(annotated_frame)

# Release everything
cap.release()
out.release()
```
